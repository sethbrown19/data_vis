## Bivariate Exploration

> In this section, investigate relationships between pairs of variables in your
data. Make sure the variables that you cover here have been introduced in some
fashion in the previous section (univariate exploration).

```python
plt.scatter(data = df, x = 'ProsperScore', y = 'BorrowerAPR');
```

We can see the apr decrease with prosper score increase. More trust in that person.

```python
sb.regplot(data = df, x = 'CreditScoreRangeUpper', y = 'EstimatedReturn', fit_reg = False,
           x_jitter = 0.2, y_jitter = 0.2, scatter_kws = {'alpha' : 1/3});
plt.figure(figsize = [12, 5]);
```

The lower the credit score the more possibilty that they are expecting a negative return.

```python
sb.violinplot(data = df, x = 'LoanStatus', y = 'ProsperScore', figsize= (15,15))
plt.xticks(rotation=90, horizontalalignment="center");
```

From this violin plot we can see where the prosper score is at and how many occurences we have of each loan status.

```python
def log_trans(x, inverse = True):
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10, x)

sb.regplot(df['OpenRevolvingMonthlyPayment'], df['BorrowerAPR'].apply(log_trans));
```

The open revolving monthly payments when receiving their APR is shown here. You would think the lower the monthly payment the better the APR but it turns out to be the opposite.

```python
sb.violinplot(data = df, x = 'LoanStatus', y = 'EsbtimatedReturn', figsize= (15,15))
plt.xticks(rotation=90, horizontalalignment="center");
```

Violin plot of loan status vs estimated return. This shows how much their estimated return they expected vs how much the loan played out nicely.

```python
def log_trans(x, inverse = True):
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10, x)

sb.regplot(df['OpenCreditLines'], df['BorrowerAPR'].apply(log_trans));
```

```python
plt.scatter(data = df, x = 'OpenCreditLines', y = 'BorrowerAPR');
```

APR vs open creditlines. there is a balance between receiving a lower APR and open credit lines.

```python
plt.scatter(data = df, x = 'OpenCreditLines', y = 'CreditScoreRangeUpper');

```

Relationship between credit score and open credit lines
