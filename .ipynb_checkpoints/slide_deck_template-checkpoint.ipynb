## Investigation Overview

> We all want to know how to have a better chance for paying off a loan and how to get the lowest APR on said loan. Prosper has made this easy with a Prosper Score. This score is based off of historical Prosper data. The score ranges from 1-10, with 10 being the best, or lowest risk score. By the end of this slide deck you will better understand what impacts this score and how having a score means you'll be more likely to pay off your loan. 

## Dataset Overview

> Prosper is a company that offers loans to human beings. The slide deck you will view below is based off of loan data provided by Prosper. This data includes things like Borrower APR, Loan Status, Employment Status, Open Credit Lines among many other categories of ranking borrowers and all the variables they use to get those rankings. 

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('C:/Users/sethb/OneDrive/Documents/Udacity_Real/data_vis/data_vis/prosperLoanData.csv')
```

## Credit Lines and Credit Score

> This scatter plot represents how much your credit score and how many open credit lines you have impacts your Prosper Score. The fewer lines of credit open correlates with a higher credit score which correlates with a hgiher prosper score.

```python
g = sb.FacetGrid(data = df, hue = 'ProsperScore', height = 5)
g.map(plt.scatter, 'OpenCreditLines', 'CreditScoreRangeUpper')
g.add_legend();
```

## Dont Be Delinquent

> The less deliquencies you have in the last 7 years the better your chances are of having a higher Prosper Score.

```python
sb.regplot(data = df, x = 'ProsperScore', y = 'DelinquenciesLast7Years', fit_reg = False,
           x_jitter = 0.2, y_jitter = 0.2, scatter_kws = {'alpha' : 1/3});
plt.figure(figsize = [12, 5]);
```

## Why High Propser Scores Matter

> As you can below, the first violin is the completed loan status. The higher you go on the prosper score, the wider the violin becomes indicating that if you have a higher prosper score, the more likely you are to pay it off. Also the violin on the far left is skewed the opposite. Ther lower your score the more likely for you to be past due on payments by over 120 days.

```python
sb.violinplot(data = df, x = 'LoanStatus', y = 'ProsperScore', figsize= (15,15))
plt.xticks(rotation=90, horizontalalignment="center");
```
