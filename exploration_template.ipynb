## Univariate Exploration

> In this section, investigate distributions of individual variables. If

```python
sorted_counts = df['BorrowerState'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90,
        counterclock = False, wedgeprops = {'width' : 0.4});
plt.axis('square');
```

The amount of loans from each state follow the state population pretty closely. The point of showing this was to see if there wereany outliers which there is in California. This could be a place to make a comparison of California interest rates vs the average across the United States. 

```python
plt.hist(data = df, x = 'BorrowerAPR');
```

What you're seeing here is the spread of the annual percetange rates awarded to borrowers. What caused the higher and lower APR? 

```python
plt.hist(data = df, x = 'EstimatedReturn');
```

What are the reasons that they are estimating a negative return on some of the loans? Are there correlations between higher returns and other variables like credit score?

```python
sorted_counts = df['CreditScoreRangeLower'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90,
        counterclock = False);
plt.axis('square');
```

```python
plt.hist(data = df, x = 'CreditScoreRangeUpper');
```

The pie graph and this histogram shows the spread and percentage of the upper and lower credit score. How much does this play a role on whether a person pays off their loan completed?

```python
base_color = sb.color_palette()[0]
pros_order = df['ProsperScore'].value_counts().index
sb.countplot(data = df, x = 'ProsperScore', color = base_color, order = pros_order);
```

This bar chart breaks out the rank of prosper score which is the score built using historical Prosper data. The score ranges from 1-10, with 10 being the best, or lowest risk score.

```python
plt.hist(data = df, x = 'LoanOriginalAmount');
```

How much are people getting loans for and does is there a correlation between completion of paying for the loan and how much the loan was?

```python
sorted_counts = df['LoanStatus'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90,
        counterclock = False);
plt.axis('square');
```

The amount of loans vs all of the status of those loans.
